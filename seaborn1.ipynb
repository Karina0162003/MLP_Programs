```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = sns.get_dataset_names()
df
```

```python
df= sns.load_dataset('glue')
df.head(4)
```

```python
sns.scatterplot(df['Model'])
```

```python
sns.lineplot(df['Year'])
```

```python
sns.lineplot(x="Year",y="Model",data=df,linestyle="-.")
```

```python
plt.figure(figsize=[12,4])
sns.barplot(x="Model",y="Year",data=df,hue="Year")
```

```python
sns.histplot(x="Model",y="Year",data=df,hue="Model",bins=30)
```

```python
sns.boxplot(x="Year",y="Model",data=df,hue="Score")
```
