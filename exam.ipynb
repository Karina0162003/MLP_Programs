```python
import matplotlib.pyplot as plt
```

```python
car_make_counts = data['Car Make'].value_counts()

fig, ax = plt.subplots(figsize=(4, 4))
ax.pie(car_make_counts, labels=car_make_counts.index, autopct="%1.0f%%", startangle=80)
ax.axis('equal')
ax.set_title('Distribution of Car Makes')

plt.show()
```

```python
car_make_counts = data['Date'].value_counts()

fig, ax = plt.subplots(figsize=(4, 4))
ax.pie(car_make_counts, labels=car_make_counts.index, autopct="%1.0f%%", startangle=80)
ax.axis('equal')
ax.set_title('Distribution of Car Year')

plt.show()
```
