```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

```python
sns.boxplot(x='sex', y='age', palette='husl', data=df,hue='who')
```

```python
sns.lineplot(x='sex', y='age', palette='husl', data=df,hue='who')
```

```python
sns.boxenplot(x='sex', y='age', palette='husl', data=df,hue='who')
```

```python
sns.barplot(x='sex', y='age', palette='husl', data=df,hue='who')
```

```python
sns.catplot(x='sex', y='age', palette='husl', data=df,hue='who')
```

```python
sns.displot(x='sex', y='age', palette='husl', data=df,hue='who')
```

```python
sns.scatterplot(x='sex', y='age', palette='husl', data=df,hue='who')
```

```python
sns.violinplot(x='class', y='alive', palette='husl', data=df,hue='who')
```
