```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("data.csv")
data
```

```python
data.head()
```

```python
from sklearn.cluster import KMeans
```

```python
kn=KMeans(n_clusters=3)
```

```python
kn.fit(data[["Age","Income"]])
```

```python
kn.predict([[54,'96000']])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
```

```python
ms.fit(data[["Age"]])
data["Age"]=ms.transform(data[["Age"]])
```

```python
data["Age"]
```

```python

ms.fit(data[["Income"]])
```

```python
data["Income"]=ms.transform(data[["Income"]])
```

```python
data["Income"]
```

```python
y_cluster=kn.fit_predict(data[["Age","Income"]])
```

```python
data["Cluster"]=y_cluster
```

```python
data.head(5)
```

```python
kn.cluster_centers_
```

```python
df1=data[data.Cluster==0]
```

```python
df2=data[data.Cluster==1]
```

```python
df3=data[data.Cluster==2]
```

```python
plt.scatter(df1.Age,df1.Income,color="red")
plt.scatter(df2.Age,df2.Income,color="blue")
plt.scatter(df3.Age,df3.Income,color="yellow")
plt.scatter(kn.cluster_centers_[:,0],kn.cluster_centers_[:,1],color="pink",marker="^")
```
