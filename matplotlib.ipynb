```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/fracture.csv')
df.tail(3)
```

```python
df.isna().sum()
```

```python
plt.figure(figsize=[7,6])
plt.scatter(df['bmd'],df['age'],color=['r'])
plt.xlabel("bmd")
plt.ylabel("Age")
plt.title("Age and bmd")
plt.show()
```

```python
plt.figure(figsize=[9,5])
plt.plot(df['fracture'],df['bmd'],linestyle="dashed",marker="*",linewidth=3)
plt.show()
```

```python
d= df['medication'].value_counts()
d.plot(kind="barh")
plt.show()
```

```python
e=df['height_cm'].value_counts()
e.plot(kind="hist",bins=50)
plt.grid(axis='y')
plt.show()
```

```python
r=df['medication'].value_counts()
```

```python
plt.pie(r,autopct="%1.1f%%",labels=r)
plt.legend()
plt.show()
```
