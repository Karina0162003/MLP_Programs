```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
categorie = ["Java","Py","c++"]
populaity = [32,34,56]
z=[99,67,88]
plt.bar(categorie,populaity,width=0.2,color="y",label="Pop") #,edgecolor="y",linewidth=4,linestyle=":",alpha=0.3
plt.bar(categorie,z,width=0.2,color="b",label="Pop2")#,edgecolor="r",linewidth=4,linestyle=":",alpha=0.3
plt.title("Marks")
plt.xlabel("Subjects",fontsize=10)
plt.ylabel("Marks")
plt.legend()
plt.show()
```

```python
x=['python','java','c','c++']
y=[85,70,76,70]
z=[30,45,80,60]
p=np.arange(len(x))
width=0.3
p1=[j+width for j in p]
plt.ylabel('Languages',fontsize=10)
plt.xlabel('Popular')
plt.title('Study')
c=['m','y','b','r']
plt.bar(p,y,width,color='r',align='edge',label="popularity")
plt.bar(p1,z,width,color='y',align='edge',label="popularity1")
plt.xticks(p+width,x)#rotation=5
plt.legend()
plt.show()
```

```python
#Scatterplot
x=[1,4,2,6,7,4,3]
y=[2,6,3,1,3,4,5]
m=[3,4,5,3,2,2,2]
# z=["r","g","b","y","r","g","r"]
z=[10,2,3,22,4,5,6]
s=[900,300,400,500,300,200,300]
plt.scatter(x,y,c=z,s=s,cmap="BrBG") #,alpha=0.4,,marker="*",edgecolor="y",linewidth=2,cmap="viridis"
plt.scatter(x,m,color="r",s=s) #new column
t=plt.colorbar()
t.set_label("ColorBar")
plt.xlabel("Day")
plt.ylabel("no")

plt.title("Scatter plot")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [2, 3, 4, 7, 4,6, 2]
y = [4, 3, 4, 4, 2,5, 2]

plt.stem(x, y, linefmt=":", markerfmt="r+", bottom=0, basefmt="g", label="Python", use_line_collection=True)

plt.legend()
plt.show()
```
