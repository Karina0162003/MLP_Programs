```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Q.1) read the dataset and print first 5 record.
df = pd.read_csv('/content/car_sales.csv')
print(df)
df.head(5)
```

```python
df.tail(5)
```

```python
#line chart
data = {'x':range(1,11),
        'y':[5,3,5,3,2,4,3,2,2,1]}
df1 = pd.DataFrame(data)
df1
plt.title("nothing")
plt.grid()
plt.plot(df1['x'],df1['y'],marker="x",color="r")
plt.xlabel('X-axis-days')
plt.ylabel('Y-axis-salesValue')
plt.legend(df1['x'])
plt.show()
```

```python
data1={'x':df['Car Year'],'y':df['Sale Price'],'z':df['Commission Earned']}
df1 = pd.DataFrame(data1)
df1
```

```python
plt.bar(df['Car Year'].head(5),df['Sale Price'].tail(5),color='r') #barh
plt.title('Price')
plt.legend(df['Car Year'])
# plt.legend(df['Sale Price'])
plt.grid()
# plt.grid(axis='x')
plt.show()
```

```python
#pie chart
datas={'a':df['Car Model'].head(5),'b':df['Sale Price'].head(5)}
df3=pd.DataFrame(datas)
plt.pie(df3['b'],labels=df3['a'],autopct='%1.1f%%',startangle=90)
# plt.pie(df['Sale Price'],labels=df['Car Model'],autopct="%1.1f%")
plt.show()
```

```python
#scatter chart
plt.scatter(df['Car Year'].head(5),df['Commission Earned'].tail(5),linewidths=5,marker=4,color="r")
plt.title("Scatter plt")
plt.xlabel("Car Year")
plt.ylabel("Commissio Eard")
plt.grid()
plt.show()
```
